## Estación de ensamblado.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t


def calcular_tiempo_permanencia():
    # Generar tiempo de llegada de una pieza
    llegada = np.random.exponential(scale=2)
    # Generar tiempo de duración de la tarea
    duracion_tarea = np.random.exponential(scale=1)

    tiempo_permanencia = llegada + duracion_tarea

    return tiempo_permanencia


num_piezas = 10000
tiempos_permanencia = []

# Simulación de tiempo de permanencia para cada pieza
for _ in range(num_piezas):
    tiempo_permanencia_pieza = calcular_tiempo_permanencia()
    tiempos_permanencia.append(tiempo_permanencia_pieza)

# Cálculo del tiempo medio de permanencia
tiempo_medio_permanencia = np.mean(tiempos_permanencia)

print("Promedio de tiempo de permanencia:", tiempo_medio_permanencia)

num_simulaciones = 1000
num_piezas_simulacion = 100
tiempos_medios_permanencia = []

# Realizar múltiples simulaciones para determinar el tiempo medio de permanencia promedio
for _ in range(num_simulaciones):
    tiempos_permanencia_simulacion = []

    # Simulación de tiempo de permanencia para cada pieza en una simulación
    for _ in range(num_piezas_simulacion):
        tiempo_permanencia_pieza = calcular_tiempo_permanencia()
        tiempos_permanencia_simulacion.append(tiempo_permanencia_pieza)

    tiempo_medio_permanencia_simulacion = np.mean(tiempos_permanencia_simulacion)
    tiempos_medios_permanencia.append(tiempo_medio_permanencia_simulacion)

# Cálculo del tiempo medio de permanencia promedio
tiempo_medio_permanencia_promedio = np.mean(tiempos_medios_permanencia)

print("Promedio del tiempo medio de permanencia:", tiempo_medio_permanencia_promedio)

# Gráfica de la estabilidad del tiempo medio de permanencia en las simulaciones
plt.plot(tiempos_medios_permanencia, color='blue')  # Cambio de color a azul
plt.xlabel('Simulación')
plt.ylabel('Promedio de tiempo de permanencia')
plt.title('Estabilidad del promedio de tiempo de permanencia en el proceso')
plt.show()

# Histograma del tiempo medio de permanencia en las simulaciones
plt.hist(tiempos_medios_permanencia, bins=30, color='green', edgecolor='orange')  # Cambio de color a verde
plt.xlabel('Promedio de tiempo de permanencia')
plt.ylabel('Frecuencia')
plt.title('Histograma del promedio de tiempo de permanencia en el proceso')
plt.show()

# Cálculo de la media y desviación estándar del tiempo medio de permanencia
media_promedio = np.mean(tiempos_medios_permanencia)
desviacion_promedio = np.std(tiempos_medios_permanencia)

print("Media del promedio de tiempo de permanencia:", media_promedio)
print("Desviación estándar del promedio de tiempo de permanencia:", desviacion_promedio)

grados_libertad = num_simulaciones - 1
media_promedio = np.mean(tiempos_medios_permanencia)
desviacion_promedio = np.std(tiempos_medios_permanencia)

# Cálculo del intervalo de confianza del tiempo medio de permanencia
intervalo_confianza = t.interval(0.95, df=grados_libertad, loc=media_promedio, scale=desviacion_promedio / np.sqrt(num_simulaciones))

print("Intervalo de confianza (95%):", intervalo_confianza)
```
