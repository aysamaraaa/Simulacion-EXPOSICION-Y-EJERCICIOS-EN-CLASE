```python
import numpy as np
import matplotlib.pyplot as plt

# Generar 15 números aleatorios entre 0 y 1
numeros_aleatorios = np.random.rand(15)

# Generar 1000 números con distribución exponencial (escala=1.0)
numeros_exponenciales = np.random.exponential(scale=1.0, size=1000)

# Crear una figura y dos subtramas
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Graficar histograma de los números aleatorios
axes[0].hist(numeros_aleatorios, bins=15, color='blue', alpha=0.7)
axes[0].set_title('Números Aleatorios')
axes[0].set_xlabel('Valor')
axes[0].set_ylabel('Frecuencia')

# Graficar histograma de la distribución exponencial
axes[1].hist(numeros_exponenciales, bins=30, color='green', alpha=0.7)
axes[1].set_title('Distribución Exponencial')
axes[1].set_xlabel('Valor')
axes[1].set_ylabel('Frecuencia')

# Ajustar el espaciado entre las subtramas y mostrar la figura
plt.tight_layout()
plt.show()

# Generar 10000 números aleatorios entre 0 y 1
numeros_aleatorios = np.random.rand(10000)

# Generar 1000 números con distribución exponencial (escala=1.0)
numeros_exponenciales = np.random.exponential(scale=1.0, size=1000)

# Crear una nueva figura y dos subtramas
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Graficar histograma de los números aleatorios
axes[0].hist(numeros_aleatorios, bins=50, color='green', alpha=0.7)
axes[0].set_title('Números Aleatorios')
axes[0].set_xlabel('Valor')
axes[0].set_ylabel('Frecuencia')

# Graficar histograma de la distribución exponencial
axes[1].hist(numeros_exponenciales, bins=50, color='blue', alpha=0.7)
axes[1].set_title('Distribución Exponencial')
axes[1].set_xlabel('Valor')
axes[1].set_ylabel('Frecuencia')

# Ajustar el espaciado entre las subtramas y mostrar la figura
plt.tight_layout()
plt.show()
```
