# Central de un servicio de emergencias

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t


def calcular_tiempo_respuesta():
    unidades = [1, 2, 3]  # Números de las unidades
    medias = [2.0, 2.1, 1.9]  # Medias de las distribuciones normales de duración
    desviaciones = [0.4, 0.3, 0.5]  # Desviaciones estándar de las distribuciones normales de duración

    duraciones_llamadas = []  # Lista para almacenar las duraciones de las llamadas

    for unidad, media, desviacion in zip(unidades, medias, desviaciones):
        duracion = np.random.normal(loc=media, scale=desviacion)  # Generar una duración aleatoria para la unidad
        duraciones_llamadas.append(duracion)

    tiempo_respuesta = max(duraciones_llamadas)  # El tiempo de respuesta será el máximo de las duraciones de las llamadas

    return tiempo_respuesta


# Realizar una corrida de simulación para determinar el tiempo de respuesta
tiempo_respuesta_simulado = calcular_tiempo_respuesta()

print("Respuesta simulada:", tiempo_respuesta_simulado)

# Realizar múltiples simulaciones del tiempo de respuesta
num_simulaciones = 10000
tiempos_respuesta_simulados = []

for _ in range(num_simulaciones):
    tiempo_respuesta_simulado = calcular_tiempo_respuesta()
    tiempos_respuesta_simulados.append(tiempo_respuesta_simulado)

# Calcular la media y desviación estándar de los tiempos de respuesta simulados
media_tiempo_respuesta = np.mean(tiempos_respuesta_simulados)
desviacion_tiempo_respuesta = np.std(tiempos_respuesta_simulados)

print("Promedio:", media_tiempo_respuesta)
print("Desviación estándar:", desviacion_tiempo_respuesta)

# Graficar la estabilidad de las simulaciones
plt.plot(tiempos_respuesta_simulados, color='blue')
plt.xlabel('Simulación')
plt.ylabel('Respuesta')
plt.title('Estabilidad en 10000 simulaciones')
plt.show()

# Graficar un histograma de los tiempos de respuesta simulados
plt.hist(tiempos_respuesta_simulados, bins=30, color='green', edgecolor='orange')
plt.xlabel('Respuesta')
plt.ylabel('Frecuencia')
plt.title('Histograma en 10000 simulaciones')
plt.show()

# Calcular nuevamente la media y desviación estándar de los tiempos de respuesta simulados
media_tiempo_respuesta = np.mean(tiempos_respuesta_simulados)
desviacion_tiempo_respuesta = np.std(tiempos_respuesta_simulados)

print("Promedio:", media_tiempo_respuesta)
print("Desviación estándar:", desviacion_tiempo_respuesta)

# Calcular el intervalo de confianza del tiempo de respuesta
grados_libertad = num_simulaciones - 1
media_tiempo_respuesta = np.mean(tiempos_respuesta_simulados)
desviacion_tiempo_respuesta = np.std(tiempos_respuesta_simulados)
intervalo_confianza = t.interval(0.95, df=grados_libertad, loc=media_tiempo_respuesta, scale=desviacion_tiempo_respuesta / np.sqrt(num_simulaciones))

print("Intervalo de confianza (95%):", intervalo_confianza)
```
