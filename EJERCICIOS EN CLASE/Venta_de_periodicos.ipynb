```python
import random
import matplotlib.pyplot as plt

lista_beneficio_promedio = []  # Creamos una lista vacía para guardar los beneficios promedio
soles = 0
aguilas = 0
num_aleatorios = 3  # Cantidad de números aleatorios generados.
dos_soles = 0
lanzamientos = 3
pedido = 50
puv = 1.10
cua = 0.90
cur = 0.10
cvp = 0.20
beneficio_total = 0

for dia in range(1, 1000):
    num_aleatorio = round(random.uniform(0, 1), 4)

    if 0 <= num_aleatorio < 0.20:
        ventas_simuladas = 30
    elif 0.20 <= num_aleatorio < 0.45:
        ventas_simuladas = 40
    elif 0.45 <= num_aleatorio < 0.70:
        ventas_simuladas = 50
    elif 0.70 <= num_aleatorio < 0.90:
        ventas_simuladas = 60
    elif 0.90 <= num_aleatorio < 1:
        ventas_simuladas = 70

    if pedido == ventas_simuladas:
        ventas_reales = ventas_simuladas
        ventas_perdidas = 0
        periodicos_no_vendidos = pedido - ventas_reales

    beneficio = (puv * ventas_reales) - (cua * pedido) - (cur * periodicos_no_vendidos) - (cvp * ventas_perdidas)
    beneficio_total += beneficio
    beneficio_promedio = beneficio_total / dia
    lista_beneficio_promedio.append(round(beneficio_promedio, 5))

plt.plot(lista_beneficio_promedio)  # Dibuja el gráfico
plt.title("Venta de periódico")  # Establece el título del gráfico
plt.xlabel("Día")  # Establece el título del eje x
plt.ylabel("Beneficio")  # Establece el título del eje y
plt.show()
```
